# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y      | X_1  |
|--------|----------|----------|
|1|$Y^{1}$ | $X_1^{1}$|
|2|$Y^{2}$ | $X_1^{2}$|
|.|.         | .        |
|.|.         | .        |
|.|.         | .        |
|m|$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def gradient_descent_1d(f_derivative, x_init, alpha, epsilon=1e-4, max_iter=1000):
    x = x_init
    for _ in range(max_iter):
        grad = f_derivative(x)
        new_x = x - alpha * grad
        if abs(new_x - x) < epsilon:
            break
        x = new_x
    return x

# Derivada de f(x) = (x - 4)^2 es f'(x) = 2(x - 4)
f_derivative = lambda x: 2 * (x - 4)

# Probamos con diferentes valores de alpha
alphas = [0.1, 0.5, 1.0]
initial_x = 0  # Punto de inicio

for alpha in alphas:
    min_x = gradient_descent_1d(f_derivative, initial_x, alpha)
    print(f"Mínimo encontrado con alpha={alpha}: x = {min_x}, f(x) = {(min_x - 4) ** 2}")

# Función F(x, y)
def F(x, y):
    return np.sin(12 * x ** 2 - 14 * y ** 2 + 3) * np.cos(2 * x + 1 - np.exp(y))

# Derivadas parciales
def grad_F(x, y):
    df_dx = 24 * x * np.cos(12 * x ** 2 - 14 * y ** 2 + 3) * np.cos(2 * x + 1 - np.exp(y)) - 2 * np.sin(12 * x ** 2 - 14 * y ** 2 + 3) * np.sin(2 * x + 1 - np.exp(y))
    df_dy = -28 * y * np.cos(12 * x ** 2 - 14 * y ** 2 + 3) * np.cos(2 * x + 1 - np.exp(y)) - np.exp(y) * np.sin(12 * x ** 2 - 14 * y ** 2 + 3) * np.sin(2 * x + 1 - np.exp(y))
    return np.array([df_dx, df_dy])

def gradient_descent_2d(f_gradient, x_init, y_init, alpha, epsilon=1e-4, max_iter=1000):
    x, y = x_init, y_init
    for _ in range(max_iter):
        grad = f_gradient(x, y)
        new_x, new_y = x - alpha * grad[0], y - alpha * grad[1]
        if np.linalg.norm([new_x - x, new_y - y]) < epsilon:
            break
        x, y = new_x, new_y
    return x, y

# Inicialización
initial_x, initial_y = 0.5, 0.5
alpha_2d = 0.01

# Encontrar el mínimo
min_x, min_y = gradient_descent_2d(grad_F, initial_x, initial_y, alpha_2d)
print(f"Mínimo encontrado para F(x, y): x = {min_x}, y = {min_y}, F(x, y) = {F(min_x, min_y)}")

# Graficar en 3D
x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = F(X, Y)

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("Superficie 3D de F(x, y)")

# Mapa de contorno
ax2 = fig.add_subplot(122)
contour = ax2.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
ax2.set_title("Mapa de contorno de F(x, y)")
plt.show()
````

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

````python
# Regresión lineal con gradiente descendente
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2 * X + 0.02 * np.random.random(100)

# Inicialización de parámetros
theta = np.random.randn(2)
alpha = 0.1
m = len(X)
X_b = np.c_[np.ones((m, 1)), X]  # Agregar columna de 1s para el término independiente

# Gradiente descendente
for _ in range(1000):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta -= alpha * gradients

print(f"Coeficientes obtenidos por gradiente descendente: {theta}")
````

````python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

````python
# Definimos la función gradiente descendente
def gradiente_descendente(theta_inicial, alpha):
    epsilon = 1E-4
    # Inicializar theta y iteraciones
    theta = theta_inicial
    iteraciones = 0
    # Lista para guardar los valores de theta y f(theta)
    theta_valores = []
    f_valores = []

    while True:
        # Definir la función f(x) = (x - 4)^2
        f_theta = (theta - 4)**2
        # Derivada de f(x) respecto a theta
        df = 2 * (theta - 4)

        # Actualizar theta con el gradiente
        theta_nuevo = theta - alpha * df

        # Guardar los valores de theta y f(theta)
        theta_valores.append(theta_nuevo)
        f_valores.append((theta_nuevo - 4)**2)

        # Parar si la magnitud del gradiente es menor que epsilon
        if abs(df) < epsilon:
            break

        # Actualizar theta
        theta = theta_nuevo
        iteraciones += 1

    print(f"Se alcanzó el mínimo después de {iteraciones} iteraciones.")
    print(f"El valor mínimo aproximado de la función es: {f_valores[-1]} en theta = {theta_valores[-1]}")
    return theta_valores, f_valores

# Probar con diferentes valores de alpha
theta_inicial = 2

# Probar para 3 valores de alpha
for alpha in [0.1, 0.2, 0.5]:
    print(f"\nProbando con alpha = {alpha}")
    theta_valores, f_valores = gradiente_descendente( theta_inicial, alpha )
    x = np.arange(0,len(f_valores))
    #plt.plot(x,  theta_valores, "k--")
    plt.plot(x, f_valores, "ro--")
    plt.xlabel('Theta')
    plt.ylabel('Iteraciones')
    plt.grid()
    plt.show()
````

````python
# Definimos la función F(x, y)
def F(x, y):
    return np.sin(12*x**2 - 14*y**2 + 3) * np.cos(2*x + 1 - np.exp(y))

# Definimos el gradiente de la función
def grad_F(x, y):

    # Derivadas parciales
    df_dx = 24*x*np.cos(12*x**2 - 14*y**2 + 3)*np.cos(2*x + 1 - np.exp(y)) - \
            2*np.sin(12*x**2 - 14*y**2 + 3)*np.sin(2*x + 1 - np.exp(y))
    df_dy = -28*y*np.cos(12*x**2 - 14*y**2 + 3)*np.cos(2*x + 1 - np.exp(y)) + \
            np.exp(y)*np.sin(12*x**2 - 14*y**2 + 3)*np.sin(2*x + 1 - np.exp(y))
    return np.array([df_dx, df_dy])

# Gradiente descendente para encontrar el mínimo de F(x, y)
def gradiente_descendente(alpha, epsilon, max_iter=1E20):
    # Inicializar valores de x, y y contadores
    x, y = 0.1, 0.1  # Valores iniciales arbitrarios
    iteraciones = 0
    valores = [(x, y)]  # Para almacenar el progreso

    while iteraciones < max_iter:
        grad = grad_F(x, y)
        # Actualizar x e y usando la regla de gradiente descendente
        x_i = x - alpha * grad[0]
        y_i = y - alpha * grad[1]

        # Comprobar si el cambio es menor que epsilon
        if np.linalg.norm([x_i - x, y_i - y]) < epsilon:
            break

        x, y = x_i, y_i
        valores.append((x, y))
        iteraciones += 1

    print(f"Mínimo encontrado después de {iteraciones} iteraciones en x = {x:.4f}, y = {y:.4f}")
    return np.array(valores)

# Crear una malla de puntos (x, y) para la gráfica 3D y el contorno
x_vals = np.linspace(-0.6, 0.6, 10)
y_vals = np.linspace(-0.6, 0.6, 10)
X, Y = np.meshgrid(x_vals, y_vals)
Z = F(X, Y)

# Probar el método de gradiente descendente
alpha = 0.01  # Tasa de aprendizaje
epsilon = 1E-4  # Criterio de parada
trayectoria = gradiente_descendente(alpha, epsilon)

# Gráfica 3D de la función, Mapa de contorno y trayectoria
fig = plt.figure(figsize=(18, 7))

# Subplot 1: Gráfica 3D
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax1.set_title('Superficie 3D de F(x, y)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('F(x, y)')

# Subplot 2: Mapa de contorno
ax2 = fig.add_subplot(132)
contour = ax2.contourf(X, Y, Z, cmap='viridis', levels=50)
plt.colorbar(contour, ax=ax2)
ax2.set_title('Mapa de contorno de F(x, y)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

# Subplot 3: Mapa de contorno con la trayectoria del gradiente descendente
ax3 = fig.add_subplot(133)
contour = ax3.contourf(X, Y, Z, cmap='viridis', levels=50)
plt.colorbar(contour, ax=ax3)
ax3.plot(trayectoria[:, 0], trayectoria[:, 1], 'r.-', label='Trayectoria del gradiente descendente')
ax3.set_title('Trayectoria del gradiente descendente')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.legend()

plt.show()
````
