```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('penguins')
df = df.rename(columns={
    'species': 'Especie',
    'island': 'Isla',
    'bill_length_mm': 'Longitud_Pico',
    'bill_depth_mm': 'Profundidad_pico',
    'flipper_length_mm': 'Longitud_Aleta',
    'body_mass_g': 'Masa',
    'sex': 'Sexo'
})
df
```

***¿Cuántas especies diferentes de pingüinos hay en el dataset?***

```python
#Punto_1
# Obtener las especies únicas y su conteo
especies = df['Especie'].unique() #Diferenciamos cuantas especies hay
conteo_especies = df['Especie'].value_counts()

print("Número de especies diferentes:",especies)
print("Conteo de especies:", conteo_especies)
```

***¿Cuál es la longitud media del pico de los pingüinos de la especie Adelie?***

```python
#punto_2
# Filtrar por especie Adelie
Adelie = df[df['Especie'] == 'Adelie']

# Calcular la longitud media del pico
Media_Pico = Adelie['Longitud_Pico'].mean()

print("La longitud media del pico de los pingüinos Adelie es:", Media_Pico ,"mm")
```

***¿Cuál es la masa corporal promedio de los pingüinos de la especie Gentoo?***

```python
#punto_3
# Filtrar por especie Adelie
Gentoo = df[df['Especie'] == 'Gentoo']

# Calcular la masa de los pinguinos
Masa = Gentoo['Masa'].mean()

print("La masa de los pingüinos Gentoo es:", Masa ,"g")
```

***¿Cuál es la relación entre la longitud del pico y la masa corporal de los pingüinos?***

```python
#Punto_4
# Crear un scatter plot
sns.scatterplot(x='Longitud_Pico', y='Masa', data=df, hue='Especie')
plt.xlabel('Longitud del pico (mm)')
plt.ylabel('Masa corporal (g)')
plt.title('Relación entre la longitud del pico y la masa corporal de los pingüinos')
plt.grid() ; plt.show()
```

***¿Cuál es la isla con mayor número de pingüinos en el dataset?***

```python
#Punto_5
Isla = df['Isla'].value_counts()
print("La isla con mayor número de pingüinos es:", Isla)
```

***¿Cuál es la desviación estándar de la longitud de la aleta de los pingüinos de la especie Chinstrap?***

***¿Cuál es la proporción de machos y hembras en la especie Adelie?***

```python
#Punto_8
Sexo = df['Sexo'].value_counts()
Machos = Sexo['Male']
Hembras = Sexo['Female']
Sex = ["Macho", "Hembra"]
Numero = [Machos, Hembras]

print("La proporción de machos en la especie Adelie es:", Machos)
print("La proporción de hembras en la especie Adelie es:", Hembras)

# Crear el gráfico de barras
plt.bar(Sex, Numero , color=['blue', 'orange'])

# Añadir etiquetas y título
plt.ylabel('Cantidad de individuos')
plt.title('Comparación del Sexo')

# Mostrar el gráfico
plt.show()
```

***¿Existe una correlación significativa entre la longitud del pico y la masa corporal en las diferentes especies de pingüinos?***

```python
from scipy.stats import pearsonr
#Punto_9
# Crear un scatter plot
sns.scatterplot(x='Longitud_Aleta', y='Masa', data=df, hue='Especie')
plt.xlabel('Longitud de la aleta (mm)')
plt.ylabel('Masa corporal (g)')
plt.title('Relación entre la longitud de la aleta y la masa corporal de los pingüinos')

# Calcular la correlación y el valor p para cada especie
# Eliminar filas con NaN
df_Nan = df.dropna()

for Especie in df_Nan['Especie'].unique():
    Especie_data = df_Nan[df_Nan['Especie'] == Especie]
    correlation, p_value = pearsonr(Especie_data['Longitud_Aleta'], Especie_data['Masa'])
    print(f"Correlación para {Especie}: {correlation}")

plt.grid() ; plt.show()
```

***¿Cuál es la isla con la mayor variación en la longitud del pico entre las tres especies de pingüinos?***

```python
#Punto_16
def diferencia_aletas_por_especie(df):
  Resultados = df.groupby("Especie")["Longitud_Aleta"].agg(["min", "max"])
  Resultados["Diferencia"] = Resultados["max"] - Resultados["min"]
  Resultados = Resultados.rename(columns={"min": "Longitud mínima", "max": "Longitud máxima"})
  return Resultados[["Longitud mínima", "Longitud máxima", "Diferencia"]]

Resultados = diferencia_aletas_por_especie(df)
print(Resultados)
```

***¿Cuál es la diferencia en masa corporal promedio entre machos y hembras de la especie Chinstrap?***

```python
#Punto_17
# Filtrar por especie Chinstrap
Chinstrap = df[df['Especie'] == 'Chinstrap']
Machos = Chinstrap[(Chinstrap["Sexo"] =='Male')]
Hembras = Chinstrap[(Chinstrap["Sexo"] =='Female')]

# Calcular la masa de los pingüinos machos y hembras
Masa_machos = Machos["Masa"].mean()
Masa_hembras = Hembras["Masa"].mean()

print("La masa promedio de los pingüinos machos es:", Masa_machos, "g")
print("La masa promedio de los pingüinos hembras es:", Masa_hembras, "g")
```

 ***¿Qué especie de pingüino tiene la longitud de aleta promedio más alta?***

```python
#Punto_18
# Filtrar por especie
Chinstrap = df[df['Especie'] == 'Chinstrap']
Adelie = df[df['Especie'] == 'Adelie']
Gentoo = df[df['Especie'] == 'Gentoo']

# Calcular la longitud de la aleta de los pingüinos
Aleta_Chinstrap = Chinstrap["Longitud_Aleta"].mean()
Aleta_Adelie = Adelie["Longitud_Aleta"].mean()
Aleta_Gentoo = Gentoo["Longitud_Aleta"].mean()


print("La longitud de la aleta promedio de los pingüinos Chinstrap es:", Aleta_Chinstrap, "mm")
print("La longitud de la aleta promedio de los pingüinos Adelie es:", Aleta_Adelie, "mm")
print("La longitud de la aleta promedio de los pingüinos Gentoo es:", Aleta_Gentoo, "mm")

Promedios = {"Chinstrap": Aleta_Chinstrap,"Adelie": Aleta_Adelie, "Gentoo": Aleta_Gentoo}
Max_Aleta = max(Promedios, key=Promedios.get)

print(f"La especie de pingüino con la longitud de aleta promedio más alta es {Max_Aleta}")
```

 ***¿Cuál es la longitud del pico más pequeña registrada en el dataset y de qué especie es?***
