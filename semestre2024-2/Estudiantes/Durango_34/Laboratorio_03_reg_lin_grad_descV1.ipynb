# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y      | X_1  |
|--------|----------|----------|
|1|$Y^{1}$ | $X_1^{1}$|
|2|$Y^{2}$ | $X_1^{2}$|
|.|.         | .        |
|.|.         | .        |
|.|.         | .        |
|m|$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

##2. Crear una función para calcular la función de coste.

````python
x = df['X_1']
y = df['Y']
m=df.shape[0] # Número de datos
theta0 = 0
theta1 = 10

# Creamos de theta inicial como vector columna
theta = np.array([theta0, theta1])

# Preparamos X como matriz (m, 2) -> Transpuesta de (2, m) a (m, 2)
ones = np.ones(m)
X = np.vstack((ones, x)).T # Ahora X es de forma (m, 2)

# Definimos el rango de theta1 que se desea probar
theta_range = np.linspace(0,2,100)
J = np.zeros(len(theta_range))  # Evitamos sobrescribir theta original

# Función para calcular la hipótesis (predicción)
def H(theta, X):
  h_ = X @ theta # Producto de matrices
  return h_

# Función para calcular el coste (función de coste)
def coste(h, y):
  J = ((1/(2*m)) * ((h - y) ** 2)).mean()
  return J

````

````python
plt.plot(x, y, 'o')
````

````python
# Bucle para calcular el costo para diferentes valores de theta1
for i, t_ in enumerate(theta_range):
  # Creamos theta como vector columna con theta0 y t_
  theta_temp = np.array([theta0, t_])
  # Calculamos la hipótesis para los datos X
  h = H(theta_temp, X)
  # Calculamos el costo para la hipótesis calculada
  J[i] = coste(h, y)
````

````python
# Encontramos el valor mínimo de la función de costo y el valor de theta correspondiente
min_cost_index = np.argmin(J) #devuelve el índice del valor mínimo en J


optimal_theta1 = theta_range[min_cost_index] #es el valor de theta1 correspondiente al costo mínimo
optimal_theta = np.array([theta0, optimal_theta1]) ##es el vector de parámetros que minimizan la función de costo.

# Calculamos la hipótesis final con los valores óptimos de theta
h_optimal = H(optimal_theta, X)

# Graficamos los datos reales y la línea de regresión
plt.scatter(x, y, label="Datos reales", color="blue")
plt.plot(x, h_optimal, label=f"Regresión (theta1 = {optimal_theta1:.4f})", color="red")

plt.xlabel("X_1")
plt.ylabel("Y")
plt.title(f"Regresión Lineal y Datos Reales (Costo mínimo: {J[min_cost_index]:.4f})")
plt.legend()
plt.show()
````

````python
# Definamos la derivada de f(x) = (x - 4)^2
def derivada_f(x):
    return 2 * (x - 4)
````

````python
#tomemos la derivada de la función, un valor inicial x, la tasa de aprendizaje alpha y el error epsilon
def gradiente_descendente(derivada_f, x_inicial, alpha, epsilon=1e-4, max_iter=1000):
  x_actual = x_inicial
  for i in range(max_iter):
    x_nuevo = x_actual-alpha*derivada_f(x_actual)
    if abs(x_nuevo-x_actual)<epsilon:
      break
    x_actual = x_nuevo
  return x_actual
````

````python
for alpha in valores_alpha:
    x_min = gradiente_descendente(derivada_f, x_inicial, alpha, epsilon)
    print(f"Con alpha = {alpha}, el valor mínimo aproximado es x = {x_min:.5f}") #con 5 cifras decimales
````

###Método del gradiente descendente para encontrar el minimo

````python
#Libreria para hallar las derivadas parciales
import autograd.numpy as np
from autograd import grad
````

````python
# Definimos nuevamente la función F(x, y) para usar autograd
def F_gradiente(xy):
    x, y = xy
    return np.sin(0.5 * x**2 - 0.25 * y**2 + 3) * np.cos(2 * x + 1 - np.exp(y))
````

````python
# Calculamos las derivadas parciales
gradiente_F = grad(F_gradiente)
````

````python
# Implementamos de manera similar al punto anterior el gradiente descendente pero para dos variables
def gradiente_descendente_2d(grad_func, xy_inicial, alpha, epsilon=1e-4, max_iter=1000):
    xy_actual = np.array(xy_inicial, dtype=float)
    for i in range(max_iter):
        gradiente = grad_func(xy_actual)
        xy_nuevo = xy_actual - alpha * gradiente
        if np.linalg.norm(xy_nuevo - xy_actual) < epsilon:
            return xy_nuevo
        xy_actual = xy_nuevo
    return xy_actual
````

````python
# Parámetros iniciales
xy_inicial = [0, 0]  # Punto inicial
alpha = 0.0001         # Tasa de aprendizaje
epsilon = 1e-4       # Umbral de convergencia
````

````python
# Encontramos el mínimo de F(x, y)
minimo = gradiente_descendente_2d(gradiente_F, xy_inicial, alpha, epsilon)
print(f"El mínimo de F(x, y) se encuentra en: x = {minimo[0]:.5f}, y = {minimo[1]:.5f}")
````

````python
# Generamos los datos
X = np.linspace(0, 1, 100)
y = 0.2 + 0.2 * X + 0.02 * np.random.random(100)

# Graficamos de los datos
plt.scatter(X, y)
plt.title("Datos generados")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
````

````python
# Definimos la función de gradiente descendente
def gradiente_descendente(X, y, alpha, epsilon=1e-6, max_iter=10000):
    # parámetros iniciales
    a = 0
    b = 0
    n = len(y)  # número de puntos
    # Gradiente descendente
    for i in range(max_iter):
        y_pred = a + b * X # predicciones actuales

        # Calculamos las derivadas parciales
        a_grad = -(2/n) * np.sum(y - y_pred)
        b_grad = -(2/n) * np.sum((y - y_pred) * X)

        # Actualizamos parámetros
        a_nuevo = a - alpha * a_grad
        b_nuevo = b - alpha * b_grad

        # Si la diferencia es menor que epsilon, detener
        if np.abs(a_nuevo - a) < epsilon and np.abs(b_nuevo - b) < epsilon:
            break

        a, b = a_nuevo, b_nuevo

    return a, b
````

````python
# Parámetros del gradiente descendente
alpha = 0.1  # Tasa de aprendizaje
epsilon = 1e-6  # Umbral de convergencia
````

````python
# Aplicamos el gradiente descendente
a_opt, b_opt = gradiente_descendente(X, y, alpha, epsilon)
print(f"Los parámetros óptimos son: a = {a_opt:.5f}, b = {b_opt:.5f}")
````

````python
# Graficamos la línea de ajuste
plt.scatter(X, y, label="Datos")
plt.plot(X, a_opt + b_opt * X, color='red', label="Ajuste lineal (GD)")
plt.title("Regresión lineal con gradiente descendente")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
````
