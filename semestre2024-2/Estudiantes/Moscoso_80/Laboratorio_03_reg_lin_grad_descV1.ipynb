# Regresión lineal:

Supongamos que tenemos un sistema en el que existe un predicor con  **m** valores de entrenamiento,  asi:   $ (x^{(1)}, y^{(1)} ) , (x^{(2)}, y^{(2)}) ... (x^{(m)}, y^{(m)})$


|Training|Y      | X_1  |
|--------|----------|----------|
|1|$Y^{1}$ | $X_1^{1}$|
|2|$Y^{2}$ | $X_1^{2}$|
|.|.         | .        |
|.|.         | .        |
|.|.         | .        |
|m|$Y^{m}$ | $X_1^{m}$  |



Podemos definir un modelo lineal  como : $h(X) = \theta_0 + \theta_1 X$ con $(\theta_0 , \theta_1)$ parámetros. Nuestro objetivo es encontrar el conjunto de puntos  $(\theta_0 , \theta_1)$ que se encuentan más "cercano" a $Y$ para cada $X$.

Para la optimización, vamos a definir la función de coste **$J(\theta_1,\theta_2 )$** para las muestras de entrenamiento como aquella que garantiza la distancia euclidiana respecto a la hipotesis planteada, así:

\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( h_{\theta} (x^{(i)})-y^{(i)})^2
\end{equation}

Para encontrar los valores  $(\theta_0 , \theta_1)$ se necesita  minimizar la función de coste, que permite obtener los valores más cercanos,  esta minimización podrá ser realizada a través de diferentes métodos el más conocido es el gradiente descendente.

![](https://github.com/hernansalinas/Curso_aprendizaje_estadistico/blob/main/Sesiones/imagenes/fig00.png?raw=true=50x)




Supongamos un modelo lineal para realizar la predicción, así nuestro modelo estará basado en la siguiente hipotesis de trabajo:

$h(X) =\theta_0 + \theta_1 X$


Reemplando el modelo anterior en la función de coste, tenemos que:



\begin{equation}
J(\theta_1, \theta_2)=\frac{1}{2m} \sum_{i=0}^m ( (\theta_0 + \theta_1 X)- y^{(i)})^2
\end{equation}

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import sympy as sy

df = pd.DataFrame(data={'X_1':np.array([0,1,2,3,4,5]),'Y':np.array([0,1,2,3,4,5])})

def calc_coste_lineal(df,coeff,intercepto):
    m = len(df.index)
    summ = sum(((intercepto + (coeff*j[0])) - j[1])**2 for i,j in enumerate(df.itertuples()))
    J = (1/(2*m))*summ
    return J
    
````

# Gradiente Descendente


Para determinar el mínimo,  a través del gradiente descendente puede ser aplicada el siguiente algoritmo:

- Proponer un número aleatorios inicial $\omega_i$
- Para descender al mínimo de la función, encontremos un  valor para el cual
el deriva de la función descenciende, asi:
\begin{equation}
\omega_{i+1} = \omega_{i} - \alpha \frac{\mathrm{d}f(\omega_i)}{\mathrm{d}\omega}
\end{equation}

donde, $\alpha$ es conocido como la tasa de aprendizaje del algoritmo.

- Evaluar $f(\omega_{i+1})$

- Iterar hasta encontrar el mínimo de la función

Siguiendo el video guia, usaremos la definición de derivada por aproximación numérica (forward difference approximation):

$$\frac{\partial f}{\partial x_i} \approx \frac{f(x + h) - f(x)}{h}$$


````python
# Hagámoslo para 2 parámetros:
#Por practicidad, usaré funciones lambda como en el video guia
import matplotlib.cm as cm
cmap = cm.plasma
fOcho = lambda x,y : np.sin(((x**2)/2)-((y**2)/4)+3)*np.cos(2*x+1-np.exp(y)) #Argumento en radianes
def grad_desc_2d(firsts,lambdaF,alpha,h,maxIter):
    _X = np.linspace(-5,5,100)
    _Y = np.linspace(-5,5,100)
    xv, yv = np.meshgrid(_X, _Y)
    _Z = np.vectorize(lambdaF)(xv,yv)
    _T = np.copy(firsts)
    minXY = np.copy(firsts)
    plt.plot(firsts[0],firsts[1],"o",c="red")
    grad = np.zeros(len(firsts))
    minimum = lambdaF(*firsts)
    for _ in range(maxIter):
        for it,th in enumerate(firsts):
            _T = np.copy(firsts)
            _T[it] = _T[it]+h
            derivadaNum = (lambdaF(*_T)-lambdaF(*firsts)) / h
            grad[it] = derivadaNum
        firsts = firsts - alpha*grad
        currentP = lambdaF(*firsts)
        if (currentP<minimum):
            minimum = currentP
            minXY = firsts
        if (_ % 25 == 0):
            color = cmap(_ / maxIter)
            plt.plot(firsts[0],firsts[1],"o",c=color)
    plt.contourf(_X,_Y,_Z,100)
    plt.plot(minXY[0],minXY[1],"w+",markersize=15)
    plt.colorbar()
    plt.show()
    return minimum, minXY

minimum,minXY = grad_desc_2d([-2,-0.5],fOcho,0.025,0.09,1000)
print(f"Minimo punto alcanzado en Z={minimum} en el punto X={minXY[0]},Y={minXY[1]}")
````

# Modelo de *machine learning*: Solución general

Un modelo general para solucionar un problema de machinine learning puede ser estructurado como sigue:


a. Eliger el modelo a emplear:
\begin{equation}
h(X,\Theta)
\end{equation}

- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:

    \begin{equation}
    h(X) = (\theta_0 + \theta_1 X)
    \end{equation}


b. Eligir la funcion de coste :
-  Metrica Ecuclidiana:
\begin{equation}
J(\Theta)=\frac{1}{2m} \sum_{i=1}^m ( h_{\theta} (X^{(i)})-y^{(i)})^2
\end{equation}
- [Lista de funciones de coste que pueden ser empleadas](https://jmlb.github.io/flashcards/2018/04/21/list_cost_functions_fo_neuralnets/)


c. Aplicar el gradiente descendente iterativamente, hasta encontrar el minimo:  

\begin{equation}
\Delta \vec{\Theta} =  - \alpha \nabla J(\Theta)
\end{equation}



- En el caso de una regresion lineal tenemos que $h(X,\Theta=(\theta_0, \theta_1))$:


\begin{equation}
\theta_0 := \theta_0 - \alpha \frac{\partial J}{\partial \theta_0}
\end{equation}

\begin{equation}
\theta_1 := \theta_1 - \alpha \frac{\partial J}{\partial \theta_1}
\end{equation}

Como debemos optimizar dos parámetros, debemos calcular el gradiente para cada una de las variables, en este caso, el intercepto y la pendiente de la recta que mejor se ajusten a los datos. Usaremos el MSE como métrica para ver cuál conjunto de parámetros es el que mejor está optimizado

````python
initParams = np.random.randn(2,1)
#X_v = np.c_[np.ones((100,1)), X]
#print((X_v.dot(initParams)-y).shape)
def gradient_descent_linear_regression(X, y, alpha, initParams, maxIterations):
    samples = len(y)
    m = initParams[0]
    b = initParams[1]
    #Se puede hacer que el ciclo vaya hasta cierto número de iteraciones o que se detenga cuando el error absoluto sea menor a cierto valor.
    #Si no se sabe como se comportará el algoritmo, lo mejor usar un número max de iteraciones
    for iteration in range(maxIterations):
        yPred = (m*X)+b
        gradientM = (-2/samples)*np.sum(X*(y-yPred))
        gradientB = (-2/samples)*np.sum(y-yPred)
        m = m-(alpha*gradientM)
        b = b-(alpha*gradientB)
        mse = np.mean((y - yPred) ** 2)
    print(f"Después de {maxIterations}, los parámetros que mejor ajustan los datos son m={m}, b={b}")
    return m, b, mse


fig, axs = plt.subplots(2,2,sharex=True,sharey=True,figsize=(10,8))

testY = lambda x,m,b: b + m*X

predM1, predB1, mse1 = gradient_descent_linear_regression(X, y, 0.1,initParams, 1000)
axs[0][0].plot(X,testY(X,predM1,predB1),c="red")
axs[0][0].scatter(X,y,s=2)
axs[0][0].title.set_text('Alpha 0.1, mse {}'.format(mse1))

predM2, predB2, mse2 = gradient_descent_linear_regression(X, y, 0.01,initParams, 1000)
axs[0][1].plot(X,testY(X,predM2,predB2),c="red")
axs[0][1].scatter(X,y,s=2)
axs[0][1].title.set_text('Alpha 0.01, mse {}'.format(mse2))

predM3, predB3, mse3 = gradient_descent_linear_regression(X, y, 0.075,initParams, 1000)
axs[1][0].plot(X,testY(X,predM3,predB3),c="red")
axs[1][0].scatter(X,y,s=2)
axs[1][0].title.set_text('Alpha 0.075, mse {}'.format(mse3))

predM4, predB4, mse4 = gradient_descent_linear_regression(X, y, 0.75,initParams, 1000)
axs[1][1].plot(X,testY(X,predM4,predB4),c="red")
axs[1][1].scatter(X,y,s=2)
axs[1][1].title.set_text('Alpha 0.75, mse {}'.format(mse4))
````

• Al tomar un alpha muy pequeño bajo un número relativamente bajo de iteraciones, el modelo no puede hacer un buen fit de los datos.

• En este caso, valores estandar de alpha como 0.1 y mayores como 0.75 dieron buenos resultados muy similares entre sí. Sorpresivamente, el alpha que mejor funcionó fue el 0.75 que en otros ajustes es probable que hiciera que las predicciones divergeran.
