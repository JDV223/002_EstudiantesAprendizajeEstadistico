# **2. Metodología**

Para abordar el objetivo de predecir los precios futuros del kWh utilizando series temporales, se empleará un enfoque sistemático basado en el análisis de series temporales. Esta es una técnica estadística que se centra en el estudio de datos recopilados a intervalos de tiempo regulares. Esta técnica es fundamental para identificar patrones subyacentes, tendencias a largo plazo y variaciones estacionales, lo que facilita la proyección de resultados futuros.

¿Cómo Funciona?

El análisis de series temporales se basa en el uso de datos que están organizados cronológicamente, permitiendo así un estudio detallado de cómo evolucionan los datos a lo largo del tiempo. Cada punto de datos está asociado a una marca de tiempo específica, lo que permite analizar cambios y patrones temporales.

El análisis de series temporales sigue un flujo de trabajo sistemático que incluye:

Durante este proceso, se generan diferentes visualizaciones de la serie para poder entender su naturaleza naturaleza (determinista o estocástica).

Una serie temporal es estocástica si presenta variabilidad no predecible y sigue un comportamiento influenciado por el azar o por factores externos desconocidos. Esto significa que, aunque pueda haber patrones, la evolución de la serie en el tiempo no puede determinarse con certeza a partir de sus valores pasados. Por tanto, utilizar una exploración gráfica de la serie respecto a sus diferentes componentes temporales puede ser una forma eficaz a la hora de identificar tendencias, patrones y variaciones estacionales, y ayuda a orientar la selección del modelo de pronóstico más adecuado.

La heterocedasticidad, estocasticidad, aleatoriedad y estacionariedad son elementos fundamentales que deben analizarse en una serie temporal, con el fin de observar su estabilidad y periodicidad.

Cuando una serie presenta una varianza y media que cambian a lo largo de las observaciones temporales, la serie es impulsada a la aleatoriedad y presenta una alta estocasticidad. Eso sucede con series que se tienden a crecer y decrecer de formas arbitrarias e impredecibles.

## **2.2 Descomposición, estacionalidad y autocorrelación**

El análisis de series temporales se basa en la identificación y descomposición de sus componentes fundamentales: tendencia, estacionalidad y residuo. Este proceso permite identificar si la serie presenta un comportamiento estacionario, aleatorio y heterocedástico, los cuales son factores fundamentales en la elección del modelo de predicción.

- **Tendencia (Trend)**: Representa la evolución general de la serie a lo largo del tiempo, eliminando las fluctuaciones a corto plazo. La tendencia es la componente de la serie temporal que captura la variación de largo plazo. Se puede modelar como una función matemática que representa el valor medio de la serie a lo largo del tiempo. Si el precio promedio del kWh muestra un aumento constante en el tiempo, se puede modelar con regresión lineal o exponencial. Si la tendencia es no lineal, se puede utilizar modelos polinomiales o técnicas de suavización.

- **Estacionalidad (Seasonal)**: Una serie estacionaria es aquella en la que sus propiedades estadísticas, como la media y la varianza, permanecen constantes en el tiempo, lo que implica que su comportamiento es predecible y que las relaciones entre observaciones en diferentes momentos son estables. Por otro lado, una serie **no estacionaria** presenta cambios en la media o la varianza a lo largo del tiempo, lo que indica que las dependencias temporales pueden variar y dificultar su modelado mediante enfoques tradicionales. (MIRAR)

- **Ruido o residuo (Residual)**: Es la parte de la serie que no puede ser explicada por la tendencia ni por la estacionalidad. Representa la variabilidad aleatoria y puede indicar si la variable tiene un alto grado de estocasticidad y heterocedasticidad.
Si la componente residual tiene gran variabilidad sin patrones claros, indica que la serie tiene un alto grado de incertidumbre y es más estocástica.

Para evaluar la estacionalidad de la serie, se realizó una descomposición aditiva utilizando la técnica de *seasonal_decompose* de la librería *Statsmodels*. En este procedimiento, la serie se descompone en sus tres componentes.


Además, no sólo es importante observar y analizar la alta volatilidad de los datos. También es necesario analizar el grado de asociación de los precios en bolsa con valores de tiempos pasados. Esto se lleva a cabo con gráficos de autocorrelación.

### **Gráficos de autocorrelación**

La autocorrelación, también conocida como dependencia secuencial, es una propiedad en la que los elementos próximos entre sí, ya sea en el espacio o en el tiempo, tienden a ser más similares entre ellos que con aquellos más distantes, simplemente debido a su cercanía.

La autocorrelación y la autocorrelación parcial son medidas de asociación entre valores de series actuales y pasadas e indican cuáles son los valores de series pasadas más útiles para predecir valores futuros.

Los **gráficos de autocorrelación** son una herramienta de uso común para comprobar la aleatoriedad del conjunto de datos, determinando autocorrelaciones de la serie con valores en intervalos de tiempo diferentes. Si dichas autocorrelaciones son cercanas a cero para todas y cada una de las separaciones de intervalos de tiempo, la serie tiene un comportamiento aleatorio, en caso contrario, autorregresivo.

* **Función de autocorrelación (ACF):** Muestra la correlación entre una serie temporal y sus rezagos (lag), es decir, qué tan relacionadas están las observaciones actuales con las anteriores.

  En una serie estacionaria, la autocorrelación debería disminuir rápidamente a medida que aumenta el número de retardos (lags), porque los valores pasados dejan de influir significativamente en los valores futuros a medida que aumenta el tiempo

* **Función de Autocorrelación Parcial (PACF):** Muestra la correlación entre la serie y un rezago específico, eliminando la influencia de los rezagos intermedios.

## **2.3 Selección de modelos**

Las redes neuronales tradicionales, como las redes neuronales feedforward, han demostrado ser poderosas aplicaciones donde se desea capturar dinámicas de series temporales autocorrelaciones y no estacionales. Sin embargo, presentan limitaciones al manejar datos secuenciales, ya que no tienen memoria de entradas anteriores. Aquí es donde entran en juego las Redes Neuronales Recurrentes (RNNs). Las RNNs están diseñadas para procesar datos secuenciales y tienen conexiones que forman ciclos dirigidos, lo que les permite mantener una "memoria" de estados anteriores y, por lo tanto, modelar dependencias temporales en los datos.

Las RNNs tradicionales tienen algunas limitaciones como problemas al entrenar RNNs con retropropagación a través del tiempo (BPTT) pues los gradientes pueden volverse extremadamente pequeños o grandes, dificultando el aprendizaje de dependencias a largo plazo. También tienen dificultades para capturar relaciones que están separadas por muchos pasos temporales.



### 2.3.1 ¿Por qué Long Short-Term Memory (LSTM)?

El modelo Long Short-Term Memory es una variante de las RNNs diseñada para superar las limitaciones mencionadas en la sección anterior. Propuestas por Hochreiter y Schmidhuber en 1997, las LSTM están equipadas con una estructura especial llamada celda LSTM, que es capaz de mantener información durante largos periodos de tiempo.

Una red de LSTM es una red neuronal recurrente (RNN) que procesa datos de entrada formando un lazo con las unidades de tiempo y actualizando el estado de la RNN. El estado de la RNN (estado interno) contiene información recordada durante todas las unidades de tiempo anteriores. En este caso, se utilizó un modelo univariado uni-step que es un modelo de predicción que utiliza una sola variable (univariado) para predecir el valor de esa misma variable en el siguiente paso temporal (uni-step).

El precio del kWh en el mercado energético presenta un comportamiento heterocedástico, no estacionario y con un grado de correlación con los eventos pasados, lo que significa que no sigue un patrón fijo y su distribución cambia con el tiempo. Por esto, el LSTM es una opción óptima para realizar la predicción pues, al ser una red neuronal profunda, puede modelar relaciones altamente no lineales, adaptándose a nuevas condiciones sin necesidad de reconfigurar manualmente parámetros como es el caso de los modelos ARIMA y SARIMA.

Inicialmente se tuvieron en cuenta otros modelos tradicionales como los mencionados anteriormente pero se hacen ineficaces en series sin periodicidad clara pues asumen que la serie es estacionaria y se basa en una combinación de promedios móviles y regresión lineal

LSTM es ideal porque:

* Recuerda información a largo plazo gracias a su arquitectura con celdas de memoria.

* Es robusto frente a ruido y tendencias, especialmente si se le dan datos correctamente transformados.

* Puede manejar relaciones complejas que otros modelos no detectan.


### 2.3.1. ¿Cómo funciona una celda LSTM?


La célula de memoria de una red LSTM está compuesta por 3 compuertas clave: una compuerta de entrada, una de olvido y una de salida. Estas compuertas regulan el flujo de información dentro de la célula de memoria, permitiendo controlar qué información se retiene y cuál se debe olvidar. Gracias a este mecanismo, las LSTM tienen la capacidad de recordar información importante en secuencias largas e ignorar elementos menos relevantes.

En las redes LSTM, además del estado oculto habitual de las redes neuronales recurrentes (RNN), se introduce un segundo estado llamado estado de la celda ($h_t$). Aquí, $h_t$ representa la memoria a corto plazo de la neurona y $c_t$ representa la memoria a largo plazo.

**Compuerta de Entrada:** Determina qué nueva información debe añadirse a la memoria, combinando la entrada actual y el estado oculto previo.

**Compuerta de Olvido:** Decide qué información obsoleta debe eliminarse de la memoria, modulando el estado previo.

**Compuerta de Salida:** Genera la salida del LSTM, basándose en la entrada actual y el estado de la memoria, produciendo un nuevo estado oculto que influye en las decisiones siguientes.


La combinación de estas tres compuertas permite que la red LSTM maneje eficazmente dependencias a largo plazo.

### 2.3.2. Modelos de potenciación del gradiente

El refuerzo del gradiente (Gradient Boosting) es un tipo de algoritmo de aprendizaje automático supervisado por conjuntos que combina múltiples aprendizajes débiles para crear un modelo final. Entrena secuencialmente estos modelos (generalmente árboles de decisión, regresiones, bosques aleatorios) colocando más pesos en las instancias con predicciones erróneas, minimizando gradualmente una función de pérdida.

La clave de Gradient Boosting es la adición secuencial de modelos, donde cada modelo posterior corrige los errores cometidos por los anteriores.

#### 2.3.2.1. Modelo XGBoost

XGBoost (Extreme Gradient Boosting) es un algoritmo de aprendizaje supervisado basado en árboles de decisión optimizados mediante Gradient Boosting. La clave de Gradient Boosting es la adición secuencial de modelos, donde cada modelo posterior corrige los errores cometidos por los anteriores.

Se destaca por su velocidad, precisión y capacidad de manejar datos faltantes, lo que lo convierte en una de las técnicas más potentes para tareas de regresión y clasificación.

\
**Funcionamiento:**

Consiste en construir una serie de árboles de decisión secuenciales para corregir errores de los modelos anteriores:

1. Se entrena un árbol de decisión inicial con los datos.

2. Cada nuevo árbol se entrena para corregir los errores del anterior (minimizando el residuo de la predicción anterior).

3. Se combinan múltiples árboles para mejorar la precisión y reducir el error.

4. El proceso continúa hasta alcanzar el número máximo de árboles o hasta que no haya mejoras significativas.

## 2.4. Arquitectura de los modelos

### 2.4.1. Transformación de los datos

LSTM es ideal porque es robusto frente a ruidos y tendencias, pero para mejorar la estabilidad de cualquier modelo predictivo es ideal transformar los datos con el fin de estacionalizar la serie y estabilizar la varianza, ya que estos presentan alta volatilidad y hace que el modelo tenga dificultad para aprender y asociar patrones. Algunas formas de transformar los datos son:

  * **Logaritmo:** Convierte la serie en una tendencia más lineal.

  * **Raíz cuadrada**.

  * **Box Cox:** Hace que los datos se ajusten mejor a una distribución normal.

### 2.4.2. Previsión directa, método de Sliding Windows

La estrategia más común y efectiva para este tipo de problema es transformar la serie temporal en un conjunto de secuencias usando una ventana deslizante (o sliding window). Con este método, se utiliza un número fijo de observaciones pasadas (la ventana de "lookback") para predecir el valor futuro.

Cada ventana entonces se construye con un número de observaciones basado en la correlación de estas con la observación que se desea predecir.

### 2.4.3. Estratificación temporal y validación cruzada

La validación cruzada tradicional no puede utilizarse ya que estos utilizan partición y estratificación seleccionando secuencias de forma aleatoria y uniforme de todo el conjunto de datos y etiquetas (`StratifiedShuffleSplit`), y esto no es conveniente pues con esta forma los datos de entrenamiento, prueba y validación no reproducen el comportamiento original de la serie.

Para abordar este problema, se usa `TimeSeriesSplit`, que preserva el orden temporal de los datos al dividirlos en segmentos crecientes. Con esta forma de validación y división de la serie se mantiene la causalidad, se evitan fugas de datos, permite evaluar la estabilidad de los modelos y simular situaciones reales.

### 2.4.4. Arquitectura de la red LSTM

* **Capas LSTM:** Se define una capa inicial LSTM de 36 neuronas, con función de activación tangente hiperbólica y activación recurrente sigmoide. A esta se le activa una regulación L2 de tipo Ridge para penalizar los parámetros de regresión, y se activa la propagación de secuencias para una capa oculta siguiente LSTM con las mismas características, pero sin necesidad de propagar secuencias.

* **Capa oculta densa:** Recibe la salida de las capas anteriores con el fin de aprender combinaciones de características de los datos entrantes. Como cada ventana de datos se construye con 24 horas, cada hora se convierte en una característica (variable aleatoria) que se desea predecir, por lo que para esta capa se establecen 24 neuronas de activación. Se establece una función de activación tangente hiperbólica y una regularización L2 de tipo Ridge.

* **Capa de salida** de una unidad que realiza regresión sobre datos continuos con la función de activación lineal ('linear').

* **Capas Batch Normalization:** Se utilizan dos entre las capas profundas con el fin de normalizar la activación de cada neurona en cada capa antes y estabilizar los valores activados antes de pasarlos a la siguiente capa.

  Con esta se acelera el entrenamiento y se mejora la estabilidad del modelo.

* **Capas Dropout:** Se incluyen seguidas de las capas Batch con el fin de apagar y encender neuronas aleatoriamente durante el entrenamiento con el fin de prevenir el sobreajuste y mejorar la capacidad de generalización del modelo.

\
**Optimización del modelo con ADAM:**

  Un optimizador se usa para ajustar los pesos en la red neuronal durante el entrenamiento, con el fin de minimizar una métrica objetivo de pérdida (loss), con base en el gradiente de la función respectiva de pérdida.

  Adam utiliza dos algoritmos clásicos de optimización utilizando dos momentos móviles de los gradientes. Estos son:

  * Momentum: 1er momento, que corresponde al momento móvil del gradiente y ayuda a suavizar y acelerar la convergencia en descensos de gradiente ruidosos.

  * RMSProp: 2do momento móvil Ajusta la tasa de aprendizaje de cada peso individualmente para evitar oscilaciones.

  Adam, por tanto, ajusta automáticamente la tasa de aprendizaje de cada peso y es muy estable en datos ruidosos, lo que lo convierte en el optimizador por excelencia en redes neuronales.

### 2.4.6. Parada temprana

Esta técnica detiene automáticamente el entrenamiento de la red neuronal cuando la pérdida en los datos de validación deja de mejorar. Esto evita que el modelo siga entrenando innecesariamente y ayuda a prevenir el sobreajuste.

Monitorea la métrica de validación o pérdida y si alguna no mejora después de cierto número de épocas (`patiente`), el entrenamiento se detiene y vuelve a los pesos donde la pérdida era mínima (`restore_best_weights=True`).

\
Los modelos de potenciación de gradiente, como XGBoost, se entrenan de forma iterativa agregando árboles de decisión de manera secuencial. La parada temprana en estos funciona similar a la de una red neuronal profunda. Se establece el número de árboles (`rounds`) a partir del cual se desea parar el entrenamiento en caso de desmejora de la métrica (`Root Mean Squared Error`) y guarda las mejores estimaciones (`save_best=True`).

## 2.5. Pronósticos

La estrategia a utilizar una vez entrenado y validado el modelo consiste en seguir la estrategia de **predicción recursiva** para realizar pronósticos a múltiples pasos en el futuro.

En este enfoque, el modelo predice un valor en el siguiente paso temporal, y luego este valor se reutiliza como entrada para predecir el siguiente, repitiendo este proceso de forma iterativa.

**Desventajas:**

Si ocurren eventos imprevistos, el modelo podría no adaptarse rápidamente debido a la dependencia en valores previos generados por el propio modelo, lo que conduce a realizar pronósticos incorrectos que se estabilizan alrededor de la media a largos plazos. Esto es, a medida que el horizonte de predicción crece, la incertidumbre se incrementa, lo que puede hacer que las predicciones sean menos confiables.

* El primer retardo tiene una autocorrelación alta, lo que indica que el precio "actual" está fuertemente influenciado por el precio inmediatamente anterior. * Tampoco se observa un patrón de oscilación en los retardos altos, lo que sugiere que la serie no exhibe ni estacionalidad débil con su versión retrasada.

* El PACF indica que la autocorrelación se disipa a partir de los lags 6 y 7, lo que sugiere una ventana de datos de 5 a 7 horas para cualquier modelo autorregresivo de predicción.

```python
#@title Descomposición de la serie de tiempo - estacionalidad entre años

def descomposicion_serie(descomposicion):
  fig, axes = plt.subplots(4,1,figsize=(10,10), sharex=True)
  # axes[0].plot(df["precio_bolsa"], label="Serie Original")
  axes[0].plot(descomposicion.observed, label="Serie Original")
  axes[0].set_title("Serie de tiempo original")
  axes[0].legend()

  axes[1].plot(descomposicion.trend, label="Tendencia",color="green")
  axes[1].set_title("Tendencia")
  axes[1].legend()

  axes[2].plot(descomposicion.seasonal, label="Estacionalidad",color="orange")
  axes[2].set_title("Estacionalidad")
  axes[2].legend()

  axes[3].plot(descomposicion.resid, label="Residuo",color="red")
  axes[3].set_title("Residuo")
  axes[3].legend()

  plt.xlabel("Fecha")
  fig.tight_layout()
  return fig
# --------------------------

df = dfo.copy()
df.set_index("datetime", inplace=True)
desc_hora = sm.tsa.seasonal_decompose(df["precio_bolsa"], model="additive", period=24*365)
fig = descomposicion_serie(desc_hora)
```

```python
''' Método para crear las secuencias recurrentes para la red '''
def crear_secuencias(datos,pasos_tiempo):
  X,y=[],[]
  for i in range(len(datos)-pasos_tiempo):  # shape(24,1)
      X.append(datos[i:i+pasos_tiempo])     # Ventanas de datos cada 24 horas
      y.append(datos[i+pasos_tiempo])       # Próximo dato a predecir: 1ra hora del siguiente día
  return np.array(X), np.array(y)
```

### Predicción

```python
#@title Arquitectura LSTM y validación del modelo

def LSTM_deep(pasos_tiempo, df, feature, neuronas, learning_rate, regl2, dropout, batch_size, ne):
  data = df[feature].values

  escalador = MinMaxScaler()
  datos_escalados = escalador.fit_transform(data.reshape(-1,1))

  X,y = crear_secuencias(datos_escalados,pasos_tiempo)

  # Aplicar TimeSeriesSplit para dividir los datos de manera secuencial (sin mezclarlos)
  tscv = TimeSeriesSplit(n_splits=5)
  splits = list(tscv.split(X))

  train_idx, test_idx = splits[-1]    # Última división

  X_entreno, X_val = X[train_idx], X[test_idx]
  y_entreno, y_val = y[train_idx], y[test_idx]

  modelo = Sequential([
      LSTM(neuronas,return_sequences=True,activation="tanh",recurrent_activation="sigmoid",kernel_regularizer=l2(regl2),input_shape=(pasos_tiempo,1)),
      BatchNormalization(),
      Dropout(dropout),
      LSTM(neuronas,return_sequences=False,activation="tanh",recurrent_activation="sigmoid",kernel_regularizer=l2(regl2)),
      BatchNormalization(),
      Dropout(dropout),
      Dense(24,activation="tanh",kernel_regularizer=l2(regl2)),
      Dense(1,activation="linear")
  ])

  optimizador = Adam(learning_rate=learning_rate)
  modelo.compile(optimizer=optimizador,loss="mse",metrics=["mae"])

  parada_temprana = tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)

  historial = modelo.fit(X_entreno,y_entreno,epochs=ne,batch_size=batch_size,validation_data=(X_val,y_val),verbose=1,callbacks=[parada_temprana])

  return modelo, historial, X_entreno, X_val, y_entreno, y_val
# ------------------------------------------------------------------------------

def validacion_LSTM(modelo, historial, X_val, y_val):
  y_pred = modelo.predict(X_val)
  mse = mean_squared_error(y_val,y_pred)
  mae = mean_absolute_error(y_val,y_pred)
  r2 = r2_score(y_val,y_pred)

  print(f"Error cuadrático medio (MSE): {mse:.6f}")
  print(f"Error absoluto medio (MAE): {mae:.6f}")
  print(f"Coeficiente de determinación (R²): {r2:.6f}")

  plt.figure(figsize=(12,6))
  plt.plot(historial.history["loss"],label="Pérdida Entrenamiento")
  plt.plot(historial.history["val_loss"],label="Pérdida Validación")
  plt.plot(historial.history["mae"],label="MAE Entrenamiento")
  plt.plot(historial.history["val_mae"],label="MAE Validación")
  plt.title("Curvas de desempeño")
  plt.xlabel("Épocas")
  plt.ylabel("Métricas")
  plt.xticks(range(0,51,5))
  plt.legend()
  plt.show()

  return y_pred
```

```python
feature = 'log_precio_bolsa'
pasos_tiempo = 24         # Ventana de datos de 24 horas
neuronas = 36
learning_rate = 0.00005
regl2 = 0.012
dropout = 0.23
batch_size = 256
ne = 40

modelo, historial, X_entreno, X_val, y_entreno, y_val = LSTM_deep(pasos_tiempo, df, feature, neuronas, learning_rate, regl2, dropout, batch_size, ne)
y_pred = validacion_LSTM(modelo, historial, X_val, y_val)
```

### Pronósticos LSTM



```python
#@title Recursive multi-step forecasting

def pronostico_LSTM(modelo, df, feature, pasos_tiempo, ventanas_futuras):
  data = df[feature].values

  escalador = MinMaxScaler()
  datos_escalados = escalador.fit_transform(data.reshape(-1,1))
  # Tomamos la última ventana y predecimos el siguiente precio
  X_input = datos_escalados[-pasos_tiempo:].reshape(1, pasos_tiempo, 1)

  predicciones = []

  # Se construyen ventanas con los nuevos valores predichos
  for _ in range(ventanas_futuras):
    pred = modelo.predict(X_input, verbose=0)[0][0]   # Hacer predicción
    predicciones.append(pred)                         # Almacenar la predicción

    # Actualizar X_input eliminando el primer valor y agregando la predicción al final
    X_input = np.append(X_input[:, 1:, :], [[[pred]]], axis=1)

  # Invertir la transformación logarítmica y la normalización
  predicciones = np.array(predicciones).reshape(-1, 1)
  predicciones = escalador.inverse_transform(predicciones)    # Desescalar
  predicciones = np.exp(predicciones)                         # Inversión del logaritmo

  return predicciones
```

```python
#@title XGBoost con TimeSeriesSplit

def model_xgboost(df, feature, pasos_tiempo, narboles, prof, learning_rate, l1, l2):
  data = df[feature].values

  escalador = MinMaxScaler()
  datos_escalados = escalador.fit_transform(data.reshape(-1,1))

  X,y = crear_secuencias(datos_escalados, pasos_tiempo)
  X = X.reshape(X.shape[0], X.shape[1])

  tscv = TimeSeriesSplit(n_splits=5)
  splits = list(tscv.split(X))

  train_idx, test_idx = splits[-1]

  X_entreno, X_val = X[train_idx], X[test_idx]
  y_entreno, y_val = y[train_idx], y[test_idx]

  # Configuración de parada temprana
  parada_temprada = xgb.callback.EarlyStopping(
      rounds=10,                  # Número de rondas sin mejora antes de detener
      metric_name="rmse",
      data_name="validation_0",   # Nombre del conjunto de evaluación
      save_best=True,             # Guardar el mejor modelo
  )

  # Configuración del modelo XGBoost
  modelo = xgb.XGBRegressor(
      n_estimators=narboles,        # Número de árboles (equivalente a las épocas)
      tree_method="hist",           # Método de construcción de árboles (histograma)
      learning_rate=learning_rate,  # Tasa de aprendizaje
      max_depth=prof,               # Profundidad máxima de los árboles
      subsample=0.8,                # Proporción de muestras usadas para entrenar cada árbol
      colsample_bytree=0.8,         # Proporción de características usadas en cada árbol
      objective='reg:squarederror', # Función de pérdida para optimizar regresión
      eval_metric=['rmse', 'mae'],
      reg_alpha=l1,               # Regularización L1 Lasso
      reg_lambda=l2,              # Regularización L2 Ridge
      gamma=0.2,                  # Reducción mínima en la pérdida necesaria para hacer una nueva partición en un nodo.
      random_state=42,
      n_jobs=-1,
      verbosity=1,
      callbacks=[parada_temprada]   # Llamado a la parada temprana
  )

  modelo.fit(X_entreno, y_entreno, eval_set=[(X_val, y_val)], verbose=True)

  return modelo, X_entreno, X_val, y_entreno, y_val
# ------------------------------------------------------------------------------

def validacion_XGBoost(modelo, X_val, y_val):
  y_pred = modelo.predict(X_val)
  mse = mean_squared_error(y_val,y_pred)
  mae = mean_absolute_error(y_val,y_pred)
  r2 = r2_score(y_val,y_pred)

  print(f"\nError cuadrático medio (MSE): {mse:.6f}")
  print(f"Error absoluto medio (MAE): {mae:.6f}")
  print(f"Coeficiente de determinación (R²): {r2:.6f}")

  return y_pred
```
